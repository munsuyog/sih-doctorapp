```python
import random
from openpyxl import Workbook

locations = [chr(i) for i in range(ord('A'), ord('N'))]
sub_locations = [f'{chr(i)}-{j}' for i in range(ord('A'), ord('N')) for j in range(1, 6)]

data_doc = {
    'Doctor Name': [f'Doctor {i}' for i in range(1, 501)],
    'Specialization': [random.choice(no_of_specialist) for _ in range(500)],
    'Location': [random.choice(locations) for _ in range(500)],
    'Sub-Location': [random.choice(sub_locations) for _ in range(500)]
}

df_random_doc = pd.DataFrame(data_doc)

df_random_doc.to_excel('doctorsdata.xlsx', index=False)
```
